```python
# Step 1: Copy dataframe for cleaning
df_clean = df.copy()
```

```python
# Step 2: Convert financial columns (remove $ , and convert to float)
def clean_currency(x):
    if isinstance(x, str):
        return float(x.replace("$", "").replace(",", "").strip())
    return x
```

```python
currency_cols = ["GDP", "Health Exp/Capita", "Tourism Inbound", "Tourism Outbound"]
for col in currency_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].apply(clean_currency)
```

```python
# Step 3: Convert Business Tax Rate to numeric (it may have % sign)
def clean_percentage(x):
    if isinstance(x, str):
        return float(x.replace("%", "").strip())
    return x
```

```python
if "Business Tax Rate" in df_clean.columns:
    df_clean["Business Tax Rate"] = df_clean["Business Tax Rate"].apply(clean_percentage)
```

```python
# Clustering Preparation

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Select useful numeric indicators

features = [
    "GDP", "Birth Rate", "Infant Mortality Rate",
    "Life Expectancy Female", "Life Expectancy Male",
    "Health Exp % GDP", "Health Exp/Capita", "CO2 Emissions",
    "Energy Usage", "Internet Usage", "Mobile Phone Usage",
    "Population Total", "Population Urban", "Business Tax Rate",
    "Days to Start Business", "Ease of Business"
]

# Keep only available features from dataset
features = [col for col in features if col in df_clean.columns]
df_features = df_clean[features]
```

```python
# Step 2: Handle missing values

# Use median imputation (robust against outliers)
imputer = SimpleImputer(strategy="median")
df_imputed = pd.DataFrame(imputer.fit_transform(df_features), columns=features)
```

```python
# Step 3: Scaling features (Standardization)

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_imputed), columns=features)
```

```python
# Step 4: Final check

print("Final dataset shape (for clustering):", df_scaled.shape)
print("\nSample records (scaled):\n", df_scaled.head())
```

```python
# Step 5: Save clean dataset for clustering

df_ready = pd.concat([df_clean["Country"], df_scaled], axis=1)
df_ready.to_csv("Cleaned_World_Development.csv", index=False)
print("\n Cleaned dataset saved as 'Cleaned_World_Development.csv'")
```

# Model Building 

```python
# 1: K-MEANS CLUSTERING MODEL

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming df_scaled and df_clean are already created
X = df_scaled.copy()
```

```python
# Elbow Method
inertia = []
K = range(2, 11)
```

```python
for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertia.append(model.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o')
plt.title("Elbow Method to Find Optimal k")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# Silhouette Score
silhouette_scores = []
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, labels))
```

```python
plt.figure(figsize=(8,5))
plt.plot(K, silhouette_scores, marker='o', color='green')
plt.title("Silhouette Score for Different k Values")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
# Fit Final K-Means Model
optimal_k = 3  # Choose based on elbow/silhouette
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df_ready["KMeans_Cluster"] = kmeans_final.fit_predict(X)
```

```python
# Visualize Clusters using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(X)
df_pca = pd.DataFrame(pca_data, columns=["PCA1", "PCA2"])
df_pca["Cluster"] = df_ready["KMeans_Cluster"]
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df_pca, x="PCA1", y="PCA2", hue="Cluster", palette="tab10", s=80)
plt.title("K-Means Clustering Visualization (PCA)")
plt.show()
```

```python
# 2: HIERARCHICAL CLUSTERING MODE

from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
# Dendrogram to visualize hierarchical relationships
plt.figure(figsize=(12,6))
linkage_matrix = linkage(X, method='ward')
dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.title("Hierarchical Clustering Dendrogram (Truncated)")
plt.xlabel("Countries")
plt.ylabel("Distance")
plt.show()
```

```python
# Apply Agglomerative Clustering
hier_cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
df_ready["Hierarchical_Cluster"] = hier_cluster.fit_predict(X)
```

```python
# Visualize PCA with Hierarchical clusters
plt.figure(figsize=(10,7))
sns.scatterplot(
    data=df_pca, x="PCA1", y="PCA2",
    hue=df_ready["Hierarchical_Cluster"], palette="Set1", s=80
)
plt.title("Hierarchical Clustering (PCA Projection)")
plt.show()
```

```python
# 3:  DBSCAN CLUSTERING MODEL

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np
```

```python
# Find best eps (distance threshold)
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
```

```python
distances = np.sort(distances[:,4], axis=0)
plt.figure(figsize=(8,5))
plt.plot(distances)
plt.title("DBSCAN - K-Distance Graph (Choose Eps)")
plt.xlabel("Data Points Sorted by Distance")
plt.ylabel("5th Nearest Neighbor Distance")
plt.show()
```

```python
# Apply DBSCAN (tune eps after viewing graph)
dbscan = DBSCAN(eps=2.5, min_samples=5)
df_ready["DBSCAN_Cluster"] = dbscan.fit_predict(X)
```

```python
# Check how many clusters and outliers (-1 = outlier)
print(df_ready["DBSCAN_Cluster"].value_counts())
```

```python
# Visualize clusters
plt.figure(figsize=(10,7))
sns.scatterplot(
    data=df_pca, x="PCA1", y="PCA2",
    hue=df_ready["DBSCAN_Cluster"], palette="rainbow", s=80
)
plt.title("DBSCAN Clustering Visualization (PCA)")
plt.show()
```

```python
# Compare cluster means for KMeans
cluster_profile = df_ready.groupby("KMeans_Cluster").mean(numeric_only=True).round(2)
sns.heatmap(cluster_profile, annot=True, cmap="YlGnBu")
plt.title("Cluster Profiles - KMeans")
plt.show()
```

```python
# CLUSTERING MODEL COMPARISON REPORT

from sklearn.metrics import silhouette_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Prepare results dictionary
model_comparison = []
```

```python
#  K-MEANS
kmeans_labels = df_ready["KMeans_Cluster"]
kmeans_score = silhouette_score(df_scaled, kmeans_labels)

model_comparison.append({
    "Model": "K-Means Clustering",
    "No. of Clusters": len(set(kmeans_labels)),
    "Silhouette Score": round(kmeans_score, 3)
})
```

```python
#  HIERARCHICAL CLUSTERING
hier_labels = df_ready["Hierarchical_Cluster"]
hier_score = silhouette_score(df_scaled, hier_labels)

model_comparison.append({
    "Model": "Hierarchical Clustering",
    "No. of Clusters": len(set(hier_labels)),
    "Silhouette Score": round(hier_score, 3)
})
```

```python
# 3: DBSCAN
dbscan_labels = df_ready["DBSCAN_Cluster"]
# Remove outliers (-1) for silhouette calculation
dbscan_mask = dbscan_labels != -1

if len(set(dbscan_labels)) > 1 and sum(dbscan_mask) > 1:
    dbscan_score = silhouette_score(df_scaled[dbscan_mask], dbscan_labels[dbscan_mask])
else:
    dbscan_score = 0  # If DBSCAN didn't form valid clusters

model_comparison.append({
    "Model": "DBSCAN",
    "No. of Clusters": len(set(dbscan_labels)),
    "Silhouette Score": round(dbscan_score, 3)
})
```

```python
# Convert to DataFrame
comparison_df = pd.DataFrame(model_comparison)

# Display results
print("\n MODEL COMPARISON REPORT:\n")
print(comparison_df)
```

```python
# Visualize the comparison

plt.figure(figsize=(8,5))
sns.barplot(data=comparison_df, x="Model", y="Silhouette Score", palette="Set2")
plt.title("Clustering Model Comparison (Silhouette Score)")
plt.ylabel("Silhouette Score")
plt.xlabel("Model")
plt.ylim(0, 1)
plt.show()
```

```python
# Display number of clusters visually
plt.figure(figsize=(8,5))
sns.barplot(data=comparison_df, x="Model", y="No. of Clusters", palette="Set3")
plt.title("Number of Clusters per Model")
plt.ylabel("No. of Clusters")
plt.xlabel("Model")
plt.show()
```
